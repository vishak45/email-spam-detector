```python
import pandas as pd


ds = pd.read_csv("email.csv", on_bad_lines='skip')


## Cleaning the data

ds.dropna(inplace=True)

ds.columns

```

```python
## Label encoder

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
ds["Category"]=le.fit_transform(ds["Category"])
print("spam,ham",ds["Category"].unique())
```

```python
## Tf-idf vector
X_raw=ds["Message"]
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer(stop_words="english",max_features=5000)
X=vectorizer.fit_transform(X_raw)

y=ds["Category"]
```

```python
## Train test split

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
## Model training and prediction

from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(X_train,y_train)

y_pred=model.predict(X_test)

from sklearn.metrics import accuracy_score

print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d',
            xticklabels=['Predicted Ham','Predicted Spam'],
            yticklabels=['Actual Ham','Actual Spam'])
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix")
plt.show()
```

```python
test="Congratulations! You have won a free trip for two to Rome. To claim your prize, give your correct bank account number. Valid 12 hours only."

data=vectorizer.transform([test])

prediction=model.predict(data)
label = le.inverse_transform([prediction])[0]

print(label)

```
